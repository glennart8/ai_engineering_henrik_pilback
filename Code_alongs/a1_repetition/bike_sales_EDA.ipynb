# Explorotory Data Analysis (EDA) on Bike Sales Dataset

```python
import pandas as pd

df = pd.read_csv("../../data/Sales.csv", index_col=0, parse_dates=True)
df.head()
```

```python
df.loc["2013"]
```

```python
df_costs = df["Unit_Cost"].sum()
df_costs
```

```python
df_price = df["Unit_Price"].sum()
df_price
```

```python
df_profit = df["Profit"].sum()
df_profit
```

```python
df_profit1 = df["Unit_Price"].sum() - df["Unit_Cost"].sum()
df_profit1
```

```python
df_canada = df[df["Country"] == "Canada"]["Profit"].sum()
df_canada
```

```python
df.head()
```

```python
df_profit_age_30plus = df[df["Customer_Age"] > 29]["Profit"].sum()
df_profit_age_30plus
```

## Hitta alla unika värden i en kolumn

```python
df_categories = df["Product_Category"].unique()
df_categories
```

## Hitta nullvärden

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

## Räknar antalet förekomster i dataframen

```python
df_top_categories = df["Product_Category"].value_counts()
df_top_categories
```

## Grupperar efter kategori och summerar profiten, descending

```python
df_profit_per_category = df.groupby("Product_Category")["Profit"].sum().sort_values(ascending=False)
df_profit_per_category
```

```python
df_by_year = df.groupby("Date")["Profit"].sum()
df_by_year
```

```python
df_by_country = df.groupby("Country")["Profit"].sum().sort_values(ascending=False)
df_by_country
```
